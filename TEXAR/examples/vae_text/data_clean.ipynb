```python
import numpy as np
```

```python
f = open('all_sent_trainVAE_less100.tsv','r')
fptb = open('simple-examples/data/ptb.train.txt','r')
```

```python
mydata = []
for line in f:
    mydata.append(line)
```

```python
ptbdata = []
for line in fptb:
    ptbdata.append(line)
```

```python
from collections import Counter

class Dictionary(object):
    def __init__(self):
        self.word2count = {}

    def add_word(self, word):
        if word not in self.word2count:
            self.word2count[word] = 1
        self.word2count[word] += 1

    def __len__(self):
        return len(self.word2count)

class Corpus(object):
    def __init__(self):
        self.dictionary = Dictionary()

    def tokenize(self, file):
        """Tokenizes a text file."""
        # Add words to the dictionary
        for line in file:
            words = line.split('\t')[0].lower().split(' ')
            for word in words:
                self.dictionary.add_word(word)
```

```python
mycorp = Corpus()
mycorp.tokenize(mydata)
myvocab = mycorp.dictionary.word2count
```

```python
ptbcorp = Corpus()
ptbcorp.tokenize(ptbdata)
ptbvocab = ptbcorp.dictionary.word2count
```
