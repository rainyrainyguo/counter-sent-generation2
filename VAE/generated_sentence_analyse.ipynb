```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Note
+ basic VAE model is trained with base VAE structure
+ RL model is trained using extra RL training, 
I trained 4 RL models with slight different epoches and random seed (some of them are trained using RL and VAE training consecutively)
+ for every model, I generate around 20000 sentences under evaluation mode, and print all the information in the log file:

log file format: **[iteration,prediction score difference, male prediction score, female prediction score, sentence]**

## basic VAE model:
### pre_yelp128_all_sent_trainVAE_sph_less100.bin

```python
# reading log file

with open('baseVAE_sph_log','r') as f:
    base = f.readlines()

difb = []
mpreb = []
fpreb = []
sentb = []
for i,x in enumerate(base[2:]):
    if i%2==0: #numbers
        y = x.split('\t')
        difb.append(float(y[1]))
        mpreb.append(float(y[2]))
        fpreb.append(float(y[3]))
    else:
        sentb.append(x.strip('\n'))

difb = np.array(difb)
mpreb = np.array(mpreb)
fpreb = np.array(fpreb)
sentb = np.array(sentb)

dfb = pd.DataFrame({"dif":difb,
                    "mpre":mpreb,
                    'fpre':fpreb,
                    "sent":sentb})
```

# histogram comparison

## RL model 1

```python
plt.figure(figsize=(15,11))
plt.hist(difb,bins=400,alpha=0.5,label='base',color='b',density=True)
plt.hist(dif,bins=400,alpha=0.3,label='RL1',color='r',density=True)
plt.legend(loc='upper right',fontsize=20)
plt.show()
```

----------------------

## some conclusions

------------------------------------------

1. the red line is **lower** for smaller number and is **higher** for bigger number, which is shifting the original sentence distribution to the one we want

```python
# look at some highly repetitively generated sentences, sort in descending number by count

df.groupby('dif')['sent'].count().reset_index(name='count').sort_values(['count'], ascending=False).head(10)
```

2. note that the red line has a very high peak on the number **3.7523856163024902**, because RL model1 sometimes repetitively generate the same sentence: **this place is closed .** (I looked up in the **training dataset** and sentence with similar meaning are among the top sentences with very different scores)
    + several peaks for generated sentences:
        - prediction_score_difference: **3.7523856163024902**; sentence:  **this place is closed.**
        - prediction_score_difference: **0.17841297388076782**; sentence:  **this place is closed . sign says coming soon detroit coney grill.**
        - prediction_score_difference: **3.7357466220855713**; sentence:  **it is closed.**
        - prediction_score_difference: **2.637481212615967**; sentence:  **closed.**

        

3. also, by looking at the generated sentences, there are a bunch of generated sentences contains the phrase: **this place is closed**, as shown below: (this may indicate the model want to combine different things with this particular expression to try if the new sentence have bigger prediction score difference)

```python
# look for generated sentences containing "closed"

for x in sent:
    if "closed" in x:
        print(x)
```

```python
# look at some generated sentences with pre_dif>3.5

gen1_big = np.arange(len(dif))[dif>3.5]
for i in gen1_big:
    print(round(dif[i],2),'\t',sent[i])
```
